## Importación de libreria

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

## Obtención de los datos

```python
datos_partidos = pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-04-10/partidos.txt', delimiter = "\t")
datos_partidos.head()
```

```python
paises_partidos = datos_partidos[(datos_partidos.equipo_1 == 'Francia') & (datos_partidos.equipo_2 == 'Bélgica')]
paises_partidos
```

## Gráfico_1

```python
#Creamos una figura 
fig = plt.figure()
#Agregamos un eje 
eje = fig.add_subplot(1, 1, 1)
#Ahora graficamos los datos de Francia y Bélgica en el mismo eje
paises_partidos.plot(x = 'anio', y ='equipo_1_final', ax = eje, linestyle = ':')
paises_partidos.plot(x = 'anio', y ='equipo_2_final', ax = eje, linestyle = '--')
#Ahora etiquetamos los ejes
plt.xlabel('Año')
plt.ylabel('Partido (resultado)')
#Titulo del grafico y leyenda
plt.title('Partidos de la copa de fútbol masculina',loc= "center",style = "italic")
plt.legend(["Francia","Bélgica"])
#guardar la figura en un archivo
plt.savefig('grafica_1.jpg')
```

De acuerdo con el *Gráfico_1* se observa que Francia ha ganado los partidos frente a Bélgica, determinando a 1986 el año con mayor anotaciones. Para Bélgica este mismo año registra el año con mayor anotaciones pero sin llegar o sobrepasar a las anotaciones de Francia.

```python
paises_partidos_2 = datos_partidos[(datos_partidos.equipo_1 == 'Alemania') & (datos_partidos.equipo_2 == 'Suecia')]
paises_partidos_2
```

## Gráfico_2

```python
#Inicializamos una figura
fig = plt.figure()
#Agregamos un eje (1 fila, 2 columnas, primer eje)
eje = fig.add_subplot(1, 2, 1)
#Graficamos los datos de Alemania en el primer eje
paises_partidos_2.plot(x = 'anio', y = 'equipo_1_final', ax = eje, ls = '-.')
#Leyenda
plt.legend(["Alemania"])
#Agregamos un segundo eje (1 fila, 2 columnas, segundo eje)
eje2 = fig.add_subplot(1, 2, 2, sharey = eje)
#Graficamos los datos de Suecia en el segundo eje
paises_partidos_2.plot(x = 'anio', y = 'equipo_2_final', ax = eje2, ls = 'dotted',color = "red")
#Agregamos etiquetas a los ejes
plt.title('Partidos de la copa de fútbol masculina(Alemania vs Suecia)',loc= "right",style = "italic")
eje.set_xlabel('Año')
eje2.set_xlabel('Año')
eje.set_ylabel('Partido (resultado)')
#Leyenda
plt.legend(["Suecia"])
#guardar la figura en un archivo
plt.savefig('grafico_2.jpg')
```

De acuerdo con el *Gráfico_2* se observa que Alemania muestra una tendencia creciente apartir de 1940 hacia los posteriores años, mientras que Suecia su tendencia creciente empieza en 2006 hacia los posteriores años. El rango de años analizados es entre 1934 a 2020.
